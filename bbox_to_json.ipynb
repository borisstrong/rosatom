```python
import os
import pandas as pd
import yaml
from pathlib import Path
from eolearn.core import EOPatch
```

# Настройки

```python
DIR = '/content/drive/MyDrive/Colab Notebooks/ЦП 2021 ФИНАЛ/files_2'
```

# Чтение файла

```python
with open(DIR+'/patches.yaml') as fh:
    read_data = yaml.load(fh)

# Print YAML data before sorting
print(read_data)
```

# Функции

```python
# Функция получения датасета карты по наименованию папки 
i = 0
def get_df(dir_name):
  result_list = []
  for yaml_el in read_data['patches']:

    for path_el in os.walk(DIR+'/'+dir_name):
      if path_el[1] == ['data', 'mask']:
        eopatch = EOPatch.load(path_el[0], lazy_loading=True)
        bbox = list(eopatch['bbox'])
        dct = {
          'uid':yaml_el['uid'],
          'lat':yaml_el['lat'],
          'lng':yaml_el['lng'],
          'x1':bbox[0],
          'y1':bbox[1],
          'x2':bbox[2],
          'y2':bbox[3]
        }
        result_list.append(dct)
  df = pd.DataFrame(result_list)
  df.head()
  df.to_csv(DIR+'/'+dir_name+'/df_for_map.csv')
  return result_list
```

```python
result_list = get_df('2021-01')
```
