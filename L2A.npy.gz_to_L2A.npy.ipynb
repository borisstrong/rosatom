# Утилита распаковки L2A.npy.gz -> L2A.npy
Проходит по папкам, ищет L2A.npy.gz файл, распаковывает его в L2A.npy, удвляет L2A.npy.gz

```python
import os
import shutil
import gzip
```

```python
DIR = '/content/drive/MyDrive/Colab Notebooks/ЦП 2021 ФИНАЛ/files_2'
```

```python
for el in os.walk(DIR):
  if el[2] == ['L2A.npy.gz']:
    gz_file = el[0] + '/L2A.npy.gz'
    npy_file = el[0] + '/L2A.npy'

    with gzip.open(gz_file, 'rb') as f_in:
      with open(npy_file, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

    os.remove(gz_file)
```
